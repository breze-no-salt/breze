```python
import cPickle
import gzip

from PIL import Image
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import theano.tensor as T

import climin.stops
import climin.initialize

from IPython.core import pylabtools
pylabtools.figsize(16, 9)
from IPython.core.display import Image as image

from breze.learn import autoencoder
#from breze.learn.rica import Rica
from breze.learn.utils import tile_raster_images
```

Helper functions
================

```python
def scale_cols_to_unit_interval(arr):
    arr = arr.copy()
    arr -= arr.min(axis=0)
    arr /= arr.max(axis=0)
    arr *= 255
    return arr.astype('uint8')
```

Prepare Data
============

We load the MNIST data. It is available at http://www.iro.umontreal.ca/~lisa/deep/data/mnist/mnist.pkl.gz.

The inputs all lie between 0 and 1. Each target is a one of k variable.

```python
datafile = 'mnist.pkl.gz'
# Load data.                                                                                                   

with gzip.open(datafile,'rb') as f:                                                                        
    train_set, val_set, test_set = cPickle.load(f)                                                       

X, Z = train_set                                                                                               
VX, VZ = val_set
TX, TZ = test_set

image_dims = 28, 28
```

Configuration
=============

```python
method = 'contractive'     # One of: basic, sparse, contractive, rica, denoising

n_features = 32
max_passes = 100

feature_dims = int(math.ceil(np.sqrt(n_features))), int(math.ceil(np.sqrt(n_features)))

batch_size = 250
hidden_transfer = 'sigmoid'
out_transfer = 'sigmoid'
loss = 'squared'
#optimizer = 'adadelta', {'steprate': 0.001, 'momentum': 0.9, 'decay': 0.9}
optimizer = 'rmsprop', {'step_rate': 0.001, 'momentum': 0.9, 'decay': 0.9}
#optimizer = 'gd', {'steprate': 0.01, 'momentum': 0.95, 'momentum_type': 'nesterov'}
#optimizer = 'lbfgs'
sparsify = False
par_std = 1e-1
tied_weights = False
```

Setup
=====

```python
losses = []

if method == 'basic':
    fe = autoencoder.AutoEncoder(X.shape[1], [n_features], 
                     hidden_transfers=[hidden_transfer], out_transfer=out_transfer,
                     loss=loss, 
                     batch_size=batch_size, optimizer=optimizer, tied_weights=tied_weights)
elif method == 'sparse':
    sparsity_loss = 'bern_bern_kl'
        
    sparsity_target = 0.01
    c_sparsity = 2.5
    #batch_size = 50000

    fe = autoencoder.SparseAutoEncoder(X.shape[1], [n_features], 
                           hidden_transfers=[hidden_transfer], out_transfer=out_transfer,
                           loss=loss, 
                           sparsity_loss=sparsity_loss,
                           c_sparsity=c_sparsity,
                           sparsity_target=sparsity_target,
                           batch_size=batch_size, optimizer=optimizer, tied_weights=tied_weights)
elif method == 'denoising':
    c_noise = .2
    noise_type = 'mask'
    batch_size = 250

    fe = autoencoder.DenoisingAutoEncoder(X.shape[1], [n_features],
                              hidden_transfers=[hidden_transfer], out_transfer=out_transfer,
                              loss=loss, 
                              c_noise=c_noise, noise_type=noise_type,
                              batch_size=batch_size, optimizer=optimizer, tied_weights=tied_weights)    

elif method == 'contractive':
    c_jacobian = 1.5
    batch_size = 250

    fe = autoencoder.ContractiveAutoEncoder(X.shape[1], [n_features], 
                                hidden_transfers=[hidden_transfer], out_transfer=out_transfer,
                                loss=loss, 
                                c_jacobian=c_jacobian,
                                batch_size=batch_size, optimizer=optimizer, tied_weights=tied_weights)    
elif method == 'rica':
    c_ica = 0.1
    rica_hidden_transfer = 'identity'
    feature_transfer = 'softabs'
    
    fe = Rica(X.shape[1], n_features, c_ica=c_ica, 
              hidden_transfer=rica_hidden_transfer, feature_transfer=feature_transfer, out_transfer=out_transfer,
              loss=loss, optimizer=optimizer)
```

```python
climin.initialize.randomize_normal(fe.parameters.data)

if sparsify:
    climin.initialize.sparsify_columns(fe.filters_in_to_hidden, sparsify)
```

Training
========

```python
batches_per_pass = float(X.shape[0]) / batch_size
max_iter = int(math.ceil(max_passes * batches_per_pass))

print "max_iter", max_iter

stop = climin.stops.AfterNIterations(max_iter)
pause = climin.stops.ModuloNIterations(batches_per_pass)

# Set up a nice printout.
keys = '#', 'loss', 'val loss'#, 'step_length'
max_len = max(len(i) for i in keys)
header = '   '.join(i.ljust(max_len) for i in keys)
print header
print '-' * len(header)


for i, info in enumerate(fe.powerfit((X,), (VX,), stop, pause)):
    losses.append((info['loss'], info['val_loss']))
    
    row = '%i' % i, '%.6f' % info['loss'], '%.6f' % info['val_loss'],# '%.6f' % info['step_length']
    print '   '.join(i.ljust(max_len) for i in row)
```
